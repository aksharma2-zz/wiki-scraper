```python
raw_html = simple_get('http://www.fabpedigree.com/james/mathmen.htm')
soupHtml = BeautifulSoup(raw_html, 'html.parser')

def get_names():
    names = set()
    raw_html = simple_get('http://www.fabpedigree.com/james/mathmen.htm')
    
    if raw_html is not None:
        soupHtml = BeautifulSoup(raw_html, 'html.parser')
        for i, li in enumerate(soupHtml.find_all('li')):
            for name in li.text.split('\n'):
                if len(name) > 0:
                    names.add(name)
                    
    return list(names)

print(get_names())
    

```

```python
def get_name_hits(name):
    url_root = 'https://xtools.wmflabs.org/articleinfo/en.wikipedia.org/{}'
    response = simple_get(url_root.format(name))
    raw_html = BeautifulSoup(response, 'html.parser')
    
    if response is not None:
        for a in raw_html.find_all('a'):
            if a['href'].find('latest-60') > -1:
                result = int(a.text.replace(',', ''))
                return result
    
    return -1
        
```

```python
print(get_name_hits('Julius Plcker'))
```

```python
def get_details():
    names = get_names()[:3]
    results = {}
    
    for name in names:
        try:
            hits = get_name_hits(name)
            if hits is None or hits == -1:
                continue
            else:
                results[name] = hits
        except:
            log_error('error encountered while processing '
                      '{}, skipping'.format(name))
            
    sorted_list = sorted(results.items(), key=lambda kv: kv[1], reverse=True)
    
    for name in sorted_list:
        print('{} has {} hits'.format(name[0], name[1]))
        
print(get_details())
```
